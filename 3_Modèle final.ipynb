# Test de l’algorithme

```python
# Import des bibliothèques

import pandas as pd
import pickle
```

```python
# Import du fichier

billets_test2 = pd.read_csv("billets_test_pavel.csv")
billets_test2
```

```python
# Charger le modèle à partir du fichier

with open('logistic_regression_model.pkl', 'rb') as file:
    estimator = pickle.load(file)

# Utiliser le modèle pour prédire sur de nouvelles données

X_new = billets_test2[['height_right', 'margin_up', 'length', 'margin_low']]
y_new_pred = estimator.predict(X_new)
billets_test2 = billets_test2.assign(target=y_new_pred)
```
